#FacetGrid

 FacetGrid allows you to create a grid of subplots based on the values of a categorical variable.

```python

df = seaborn.load_dataset("tips")

#_-_-_-_-_-_- MAIN SECTION -_-_-_-_-_-_-_-_#

# Form a facetgrid using columns with a hue
graph = seaborn.FacetGrid(df, col = "sex", hue = "smoker")

# map the above from facetgrid with some attributes
graph.map(plt.scatter, "total_bill", "tip", edgecolor = "w").add_legend()

plt.show()
```

```python

tips = sns.load_dataset("tips")

g = sns.FacetGrid(tips, col = "sex", row = "time")

g.map(sns.scatterplot, "total_bill", "tip")

plt.show()
```

```python

iris = sns.load_dataset("iris")

g = sns.PairGrid(iris, hue = "species")

g.map_lower(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)

g.add_legend()

plt.show()
```

```python

df = seaborn.load_dataset("tips")

graph = seaborn.PairGrid(df, hue = "day")
graph = graph.map_diag(plt.hist)
graph = graph.map_offdiag(plt.scatter)
graph = graph.add_legend()

plt.show()
```

# Min - Max Scaling Visualization

```python

# Standardized Total Bill

from sklearn.preprocessing import MinMaxScaler

tips = sns.load_dataset("tips")

scaler_minmax = MinMaxScaler()
tips["total_bill_MinMax"] = scaler_minmax.fit_transform(tips[["total_bill"]])

sns.histplot(tips["total_bill_MinMax"], kde = True)
```

```python

# Original Total Bill

tips = sns.load_dataset("tips")

scaler_minmax = MinMaxScaler()
tips["total_bill_MinMax"] = scaler_minmax.fit_transform(tips[["total_bill"]])

sns.histplot(tips["total_bill"], kde = True)
```

# Standardization 

* Z - Score normalization.
* Tranforms the data to have a mean of 0 and a standard deviation of 1.


```python

# Standardized Total Bill

from sklearn.preprocessing import StandardScaler

tips = sns.load_dataset("tips")

scaler_standard = StandardScaler()
tips["total_bill_standard"] = scaler_standard.fit_transform(tips[["total_bill"]])

sns.histplot(tips["total_bill_standard"], kde = True)
```

```python

# Original Total Bill

tips = sns.load_dataset("tips")

scaler_standard = StandardScaler()
tips["total_bill_standard"] = scaler_standard.fit_transform(tips[["total_bill"]])

sns.histplot(tips["total_bill"], kde = True)
```


# Log Transformation Visualization

* Converts data to a logarithmic scale.
* Reduces skewness by compressing large values.


```python

tips = sns.load_dataset("tips")

log_transformed = np.log(tips["total_bill"] + 1)

sns.histplot(log_transformed, kde = True)
```


# Univariate KDE Plot

* Visualizing the distribution of a single variable (Eg : "sepal length (cm)")


```python

df = sns.load_dataset("iris")
```

```python

plt.figure(figsize = (8, 6))
sns.kdeplot(data = df, x = "sepal_length", fill = True, alpha = 0.5, hue = "species")

plt.title("UNIVARIATE KDE PLOT FOR SEPAL LENGTH")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Density")

plt.show()
```


# Bisvariate KDE Plot

* Visualizing the joint distribution of two variables (Eg : "sepal length (cm)" and "sepal width (cm)")


```python

plt.figure(figsize = (8, 6))
sns.kdeplot(data = df, x = "sepal_length", y = "sepal_width", fill = True, alpha = 0.5, hue = "species")

plt.title("BIVARIATE KDE PLOT")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")

plt.show()
```
